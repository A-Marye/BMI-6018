```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
diabetic_data = pd.read_csv("diabetic_data.csv")
diabetic_data.head(15)
```

### Melt

```python
melt_df = diabetic_data.melt(id_vars = ['age'], value_vars = ['time_in_hospital'])
melt_df.head(15)
```

### Pivot

```python
pivot_df = diabetic_data.pivot(index = 'encounter_id', columns = 'gender', values = 'time_in_hospital')
pivot_df.head(15)
```

### Aggregation

```python
aggregation_df = diabetic_data.aggregate({'time_in_hospital' : ['min', 'max']})
aggregation_df.head(15)
```

### Iteration

```python
diabetic_data_short = diabetic_data[:10]
for row_index,row in diabetic_data_short.iterrows():
   print(row_index,row)
```

### Groupby

```python
groupby_df = diabetic_data.groupby('gender')['time_in_hospital'].mean()
groupby_df.head()
```
